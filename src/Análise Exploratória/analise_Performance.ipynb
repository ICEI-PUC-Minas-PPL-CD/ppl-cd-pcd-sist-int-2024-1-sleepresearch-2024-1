# Analise da tabela de Performance

## Importações

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

### Tabela

```python
data = pd.read_csv('..\\assets\\data\\Student_Performance.csv')
```

## Estatistica simples

```python
data.describe()
```

### Previos Scores

#### Alunos que tiveram a maior nota anterior

```python
MaiorPS = data[data['Previous Scores']== data['Previous Scores'].max()]
```

```python
MaiorPS.describe()
```

```python
MaiorPS['Extracurricular Activities'].value_counts()
```

#### Alunos que tiveram o menor valor de PS

```python
MenorPS = data[data['Previous Scores']== data['Previous Scores'].min()]
```

```python
MenorPS.describe()
```

```python
MenorPS['Extracurricular Activities'].value_counts()
```

#### Notas maiores que 80

```python
OitantaPS = data[data['Previous Scores']>= 80]
```

```python
OitantaPS.describe()
```

```python
OitantaPS['Extracurricular Activities'].value_counts()
```

### Extracurricular Activities

### Performance Index

```python
data['Performance Index'].median()
```

#### Performance acima da mediana

```python
Performance_AcimaMediana = data[data['Performance Index'] >= 55]
Performance_AcimaMediana.describe()
```

```python
Performance_AcimaMediana['Extracurricular Activities'].value_counts()
```

##### Tempo de estudo

```python
Performance_AcimaMediana[Performance_AcimaMediana['Hours Studied'] == 1].describe()
```

```python
Performance_AcimaMediana[Performance_AcimaMediana['Hours Studied'] == 9].describe()
```

#### Performance abaixo da mediana

```python
Performance_AbaixoMediana = data[data['Performance Index'] < 55]
Performance_AbaixoMediana.describe()
```

```python
Performance_AbaixoMediana['Extracurricular Activities'].value_counts()
```

## Gráficos

### graficos gerais

```python
g = sns.PairGrid(data)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.lineplot)
```

### Performance index

```python
sns.histplot(data['Performance Index'])
```

### Tempo de estudo

```python
ax = sns.lineplot(data=data, x= 'Hours Studied', y='Performance Index', errorbar=None)
ax.set_title('Performance X Horas de estudo')
```

```python
ax.figure.savefig("PerformanceXhours.png")
```

#### Performance Acima mediana

```python
ax = sns.lineplot(data=Performance_AcimaMediana, x= 'Hours Studied', y='Performance Index')
```

```python
ax.figure.savefig("PerformanceXhours.png")
```

#### Abaixo Mediana

```python
sns.lineplot(data = Performance_AbaixoMediana,x= 'Hours Studied', y='Performance Index')
```

#### Distribuicao

```python
newdata = pd.DataFrame({'Alunos': ["Acima da mediana (x>=55)", "Abaixo da mediana (x<55)"], 'Tempo medio de estudo': [Performance_AcimaMediana['Hours Studied'].mean(), Performance_AbaixoMediana['Hours Studied'].mean()]})
```

```python
ax = sns.barplot(data=newdata, x='Alunos', y='Tempo medio de estudo')
ax.set_title("Análise da mediana")
```

```python
ax.figure.savefig("AnaliseMediana.png")
```
