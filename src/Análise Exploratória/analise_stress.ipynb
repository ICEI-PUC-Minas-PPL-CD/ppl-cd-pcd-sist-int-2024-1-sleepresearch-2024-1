## Gráficos

### geral

```python
labels = 'sono elevado e estudo elevado', 'sono elevado e estuda baixo', 'sono baixo e estudo elevado', 'sono baixo e estudo baixo', 'sono médio e estudo médio'
```

```python
g = sns.PairGrid(data)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.lineplot)
```

### Pizza: distribuição da quantidade de pessoas em cada categoria de performance, sono e estudo

```python
SS1 = AltoSono_AltoEstudo.count()
SS2 = AltoSono_BaixoEstudo.count()
SS3 = BaixoSono_AltoEstudo.count()
SS4 = BaixoSono_BaixoEstudo.count()
SS5 = MedioSono_MedioEstudo.count()
```

```python
soma =  SS1 + SS2 + SS3 + SS4 + SS5
```

```python
sizes = [SS1,SS2,SS3,SS4,SS5]
```

```python
fig, ax1= plt.subplots()

ax1.pie(sizes, autopct='%1.1f%%', shadow=False, startangle=90)
ax1.legend(labels=labels, bbox_to_anchor = (1,1))
```

### Distribuição da performance pela media de cada categoria

```python
medias = [AltoSono_AltoEstudo.mean(), AltoSono_BaixoEstudo.mean(), AltoSono_AltoEstudo.mean(), BaixoSono_BaixoEstudo.mean(), MedioSono_MedioEstudo.mean()]
```

```python
graph1 = pd.DataFrame({'categorias': labels, 'media da performance': medias})
```

```python
ax2 = sns.barplot(data= graph1, x='categorias',y='media da performance')
ax2.set_xticklabels(['1','2', '3', '4', '5'])
ax2.legend(labels=labels, bbox_to_anchor = (1.1,1))
```

### Distribuição da média do Stress para cada categoria de Sono x Dor de cabeça x Stress

```python
mediasStress = [AltoSono_AltoEstudo.mean(), AltoSono_BaixoEstudo.mean(), BaixoSono_AltoEstudo.mean(), BaixoSono_BaixoEstudo.mean(), MedioSono_MedioEstudo.mean()]
```

```python
graph2 = pd.DataFrame({'categorias': labels, 'Stress Medio': mediasStress})
```

```python
ax3 = sns.barplot(data=graph2, x= 'categorias', y= 'Stress Medio')
ax3.set_xticklabels(['1','2', '3', '4', '5'])
ax3.legend(labels=labels, bbox_to_anchor = (1.1,1))
```
